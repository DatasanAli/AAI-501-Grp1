```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create dataframe with performance metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
models = ['XGBoost', 'ANN', 'Naive Bayes']

data = {
    'Metric': metrics * 3,
    'Model': [model for model in models for _ in range(len(metrics))],
    'Value': [
        81.7, 64.0, 38.0, 48.0,  # XGBoost
        78.0, 61.0, 32.0, 42.0,  # ANN
        72.0, 53.0, 29.0, 37.0   # Naive Bayes
    ]
}

df = pd.DataFrame(data)
```

```python
# Create the visualization
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
sns.set_palette("husl")

# Create grouped bar chart
bar_plot = sns.barplot(x='Metric', y='Value', hue='Model', data=df)

# Customize the plot
plt.title('Model Performance Comparison', fontsize=14, pad=20)
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Add value labels on top of each bar
for container in bar_plot.containers:
    bar_plot.bar_label(container, fmt='%.1f%%', padding=3)

# Adjust legend position and size
plt.legend(title='Models', bbox_to_anchor=(1.05, 1), loc='upper left')
```
